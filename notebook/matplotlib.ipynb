# Matplotlib

[Matplotlib](https://matplotlib.org/) is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms.

Use `import` to load Matplotlib.

```python
%matplotlib notebook
import matplotlib.pyplot as plt

x = list(range(0, 11))
y = list(range(0, 11))

# line plot
plt.plot(x, y)
plt.show()
```

Scatter plot

```python
x = [1, 10, 100, 1000, 10000, 100000]
y = [1, 10, 100, 1000, 10000, 100000]
plt.scatter(x, y, c = ['red', 'green', 'yellow', 'blue', 'orange', 'purple'], s = [1, 10, 20, 40, 80, 160])
plt.xscale("log")
plt.yscale("log")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("x vs. y")
plt.text(110, 110, 'Yellow')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import random
random.seed(1984)
mu, sigma = 0, 0.1 # mean and standard deviation
dist = np.random.normal(mu, sigma, 1000)

# default is 10 bins
plt.hist(dist, bins = 25)
plt.show()
```
