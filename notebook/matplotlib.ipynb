# Matplotlib

[Matplotlib](https://matplotlib.org/) is a comprehensive Python library for creating static, animated, and interactive visualisations. Matplotlib can:

* Create publication quality plots.
* Make interactive figures that can zoom, pan, update.
* Customise visual style and layout.
* Export to many file formats.
* Embed in JupyterLab and Graphical User Interfaces.

[Pyplot](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.html) is a Matplotlib module that provides functions for adding plot elements. Use `import` to load pyplot from Matplotlib.

```python
from matplotlib import pyplot as plt
```

## Basic plots

A simple line plot.

```python
x = list(range(0, 11))
y = list(range(10, 21))

plt.plot(x, y)
plt.show()
```

Bar plot. An axes object is added using `add_axes()`, where the arguments are for `left`, `right`, `width`, and `height`.

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 2, 1])
x = ['one', 'two', 'three', 'four', 'five']
y = [19, 84, 0, 3, 13]
ax.bar(x, y)
plt.title("A bar plot")
plt.show()
```

Scatter plot using [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) with a modified scale, x and y axis labelling, a title, text, and a grid. The `c` parameter provides the colours for each point and the `s` parameter provides the point sizes.

```python
x = [1, 10, 100, 1000, 10000, 100000]
y = [1, 10, 100, 1000, 10000, 100000]

plt.scatter(
    x,
    y,
    c = ['red', 'green', 'yellow', 'blue', 'orange', 'purple'],
    s = [1, 10, 20, 40, 80, 160]
)

plt.xscale("log")
plt.yscale("log")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("x vs. y")

plt.text(110, 110, 'Yellow')

plt.grid(True)

plt.show()
```

Histogram.

```python
import numpy as np
import random
random.seed(1984)

# mean and standard deviation
mu, sigma = 0, 0.1

dist = np.random.normal(mu, sigma, 1000)

# default is 10 bins
plt.hist(dist, bins = "auto")
plt.show()
```

## Subplots

Use `subplot()` to plot multiple plots together; the arguments are `nrows`, `ncols`, `index`.

```python
plt.subplot(2, 2, 1)
plt.plot(range(0,11), range(0,11))
plt.title("First plot")

plt.subplot(2, 2, 2)
plt.hist(dist, bins = 20)
plt.title("Second plot")

plt.subplot(2, 2, 3)
plt.plot(range(0,11), range(0,11))
plt.title("Third plot")

plt.subplot(2, 2, 4)
plt.hist(dist, bins = 20)
plt.title("Fourth plot")

plt.subplots_adjust(
    wspace=0.3,
    hspace=0.5
)

plt.show()
```
